Removemos las columnas que de plano, no utilizaremos para un correcto análisis de datos. 

```python
data = data.drop(['Unnamed: 22','Unnamed: 23','href','href formula','pdf'], axis=1)
data.head()
```

Nos percatamos que Case Number, es un formato fecha, dividido por ".". También que tenemos el año por separado. Así como otras dos columnas que marcan el case number 1 y case number 2. Para decidir si retiraremos estas últimas dos, las compararemos para saber si realmente es una réplica del case number.

```python
data = data.drop(['Case Number', 'Year'], axis=1)
data.head()
```

```python
data_01 = data['Case Number.1'] == data['Case Number.2']
data_01
```

```python
data = data.drop(['Case Number.1', 'Case Number.2'], axis=1)
data.head()
```

```python
# Transformamos la columna de original order a int
# Primero, removeremos todos los non-numeric values usando to_numeric
# Con esto reemplazaremos los non-numeric values to NaN's
data['original order'] = pd.to_numeric(data['original order'], errors='coerce')
data.head()
```

```python
# Y removemos todas las filas con valores NaN en la columna original order
data = data.dropna(subset=['original order'])
# Por último convertimos los vamos a int
data['original order'] = data['original order'].astype(int)
```

```python
data.head()
```

```python
# Renombraremos la columna Type, para que sea aún más específica
data.rename(columns = {'Type':'Attack type'}, inplace = True)
data.head()
```

```python
# Así como también la columna de Area a Estado
data.rename(columns = {'Area':'State'}, inplace = True)
data.head()
```

```python
# Y Name a Victim's Name
data.rename(columns = {'Name':'Victim Name'}, inplace = True)
data.head()
```
